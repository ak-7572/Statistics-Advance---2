```python
# Q1. Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?


### Z-Statistic and its Relationship to the Standard Normal Distribution

# The **z-statistic** (also called the **z-score**) is a measure of how many standard deviations a data point (or sample mean) is away from the population mean. It is used to standardize scores on the normal distribution and is a fundamental concept in statistics.

# The formula for the z-statistic is:

# z = \frac{X - \mu}{\sigma}


# Where:
# - \(X\) is the data point (or sample mean in the case of a sample),
# - \(\mu\) is the population mean,
# - \(\sigma\) is the population standard deviation.

# For a sample, when the population standard deviation is unknown, the z-statistic can be approximated by substituting the sample standard deviation \(s\) for \(\sigma\), especially when the sample size is large.

# The z-statistic transforms data into a form that corresponds to the **standard normal distribution**. This distribution is a normal distribution with:
# - A mean of 0,
# - A standard deviation of 1.

# Thus, the z-statistic tells you how far away \(X\) (or a sample mean) is from the population mean in terms of the number of standard deviations, and it maps to the standard normal distribution for easy comparison.

### Z-Statistic and Hypothesis Testing

# In hypothesis testing, the z-statistic is used to determine whether to reject the null hypothesis. Hereâ€™s how it works:

# 1. **State the hypotheses:**
#   - Null hypothesis (\(H_0\)): This typically represents a statement of "no effect" or "no difference."
#   - Alternative hypothesis (\(H_1\)): This represents what you aim to prove or investigate.

# 2. **Choose a significance level** (\(\alpha\)): Common values are 0.05, 0.01, or 0.10. This is the probability threshold below which you will reject the null hypothesis.

# 3. **Calculate the z-statistic**: Using the sample data and the population parameters (or sample estimates), calculate the z-statistic based on the formula mentioned earlier.

# 4. **Find the p-value**: The p-value is the probability that the observed result (or one more extreme) would occur if the null hypothesis were true. For a two-tailed test, this is the area under the standard normal curve that is beyond the z-statistic in both directions.

# 5. **Compare the p-value with \(\alpha\)**:
#   - If the p-value is less than or equal to \(\alpha\), reject the null hypothesis.
#   - If the p-value is greater than \(\alpha\), do not reject the null hypothesis.
```

```python
# Q2. What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?

### What is a p-value?

# A **p-value** (probability value) is a measure used in statistical hypothesis testing to help determine the strength of the evidence against the null hypothesis. It is the probability of obtaining a test statistic (such as a z-statistic, t-statistic, or chi-square statistic) at least as extreme as the one observed in the sample data, assuming that the null hypothesis is true.

# In simpler terms, the p-value tells you how likely it is to observe the data (or something more extreme) if the null hypothesis were correct. The smaller the p-value, the stronger the evidence against the null hypothesis.

### How is the p-value used in hypothesis testing?

# In hypothesis testing, the p-value is compared to a predefined significance level (denoted as \(\alpha\)), which is typically 0.05, 0.01, or 0.10. The decision rule based on the p-value is as follows:

# - **If the p-value is less than or equal to \(\alpha\)** (i.e., \(p \leq \alpha\)): Reject the null hypothesis (\(H_0\)).
#  - This suggests that the observed data is inconsistent with the null hypothesis, and there is sufficient evidence to support the alternative hypothesis (\(H_1\)).

# - **If the p-value is greater than \(\alpha\)** (i.e., \(p > \alpha\)): Fail to reject the null hypothesis.
#  - This means that the observed data is consistent with the null hypothesis, and there is insufficient evidence to support the alternative hypothesis.

### What does it mean if the p-value is very small (e.g., 0.01)?

# A **very small p-value** (such as 0.01) indicates strong evidence against the null hypothesis. Here's what it implies:

# - **If \(p = 0.01\)**: This means that there is only a 1% chance of obtaining a result as extreme as the one observed (or more extreme) if the null hypothesis were true. Since this probability is very small, the observed result is unlikely to occur by random chance, suggesting that the null hypothesis is probably not true.

# - In this case, if your chosen significance level \(\alpha\) is 0.05, the p-value of 0.01 would be **less than \(\alpha\)**, so you would **reject the null hypothesis**. This suggests that there is significant evidence in favor of the alternative hypothesis.

# However, it's important to note that **rejecting the null hypothesis** does not prove the alternative hypothesis is true. It simply suggests that the data provides strong evidence against the null hypothesis.

```

```python
# Q3.  Compare and contrast the binomial and Bernoulli distributions.

### Comparison and Contrast: Binomial and Bernoulli Distributions

# The **binomial distribution** and the **Bernoulli distribution** are both discrete probability distributions that describe the outcomes of experiments involving binary (two possible) outcomes. However, they differ in scope and application. Let's break down their key similarities and differences:

### 1. **Definition**

#  - **Bernoulli Distribution**:
#  - The Bernoulli distribution models a **single trial** or experiment with two possible outcomes: success (often denoted by 1) and failure (often denoted by 0).
#  - It describes the probability of success in a single trial.
#  - The distribution has only one parameter: the probability of success, \( p \), where \( 0 \leq p \leq 1 \).


#  P(X = 1) = p \quad \text{(probability of success)}
#  P(X = 0) = 1 - p \quad \text{(probability of failure)}


# - **Binomial Distribution**:
#  - The binomial distribution generalizes the Bernoulli distribution to **multiple trials**. It models the number of successes in a fixed number of independent trials, each with the same probability of success, \( p \).
#  - It has two parameters: \( n \) (the number of trials) and \( p \) (the probability of success in each trial).

#  The probability mass function for a binomial distribution is:

#   P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \quad \text{for} \ k = 0, 1, \dots, n


#  Where \( k \) is the number of successes, and \( \binom{n}{k} \) is the binomial coefficient, representing the number of ways to choose \( k \) successes from \( n \) trials.

### 2. **Number of Trials**

# - **Bernoulli Distribution**: It models only **one trial** (a single experiment).
# - **Binomial Distribution**: It models **multiple trials** (specifically, a fixed number of trials, \( n \)).

### 3. **Parameters**

# - **Bernoulli Distribution**: It has **one parameter**, \( p \), which is the probability of success in a single trial.

# - **Binomial Distribution**: It has **two parameters**: \( n \) (the number of trials) and \( p \) (the probability of success in each trial).

### 4. **Outcomes**

# - **Bernoulli Distribution**: The outcome is binary, either 0 (failure) or 1 (success).

# - **Binomial Distribution**: The outcome is a count of successes over \( n \) trials, so the possible outcomes are any integer between 0 and \( n \) (i.e., \( 0, 1, 2, \dots, n \)).

### 5. **Mean and Variance**

# - **Bernoulli Distribution**:
# - The **mean** (expected value) is \( E(X) = p \).
# - The **variance** is \( \text{Var}(X) = p(1 - p) \).

# - **Binomial Distribution**:
#  - The **mean** (expected value) is \( E(X) = n \cdot p \).
#  - The **variance** is \( \text{Var}(X) = n \cdot p \cdot (1 - p) \).

### 6. **Relation Between the Two Distributions**

# The **binomial distribution** is a generalization of the **Bernoulli distribution**. In fact, if we set \( n = 1 \) in the binomial distribution, we get the Bernoulli distribution. This means that:
# - A **single trial** (Bernoulli trial) is a special case of the **binomial distribution** where the number of trials is one (\( n = 1 \)).

### 7. **Applications**

# - **Bernoulli Distribution**:
#  - Used when modeling a single trial or experiment with two possible outcomes. For example, flipping a coin once (heads or tails), or testing whether a product is defective or not in a single inspection.

# - **Binomial Distribution**:
#  - Used when modeling the number of successes in a fixed number of independent trials. For example, flipping a coin 10 times and counting how many times you get heads, or conducting 100 product inspections and counting how many are defective.
```

```python
# Q4.  Under what conditions is the binomial distribution used, and how does it relate to the Bernoulli distribution?


### Conditions for Using the Binomial Distribution

# The **binomial distribution** is used under the following conditions:

# 1. **Fixed Number of Trials**:
#   - The experiment involves a **fixed number of trials**, denoted as \( n \). For example, you might flip a coin 10 times, or perform 20 independent trials.

# 2. **Binary Outcomes**:
#   - Each trial must result in one of two possible outcomes, often referred to as "success" and "failure". For example, in a coin flip, the two outcomes are heads (success) and tails (failure), or a product is defective (failure) or not (success).

# 3. **Independence of Trials**:
#   - The trials must be **independent**, meaning the outcome of one trial does not affect the outcome of any other trial. For example, the result of one coin flip doesn't influence the next flip.

# 4. **Constant Probability of Success**:
#   - The probability of success (denoted by \( p \)) must be the same for each trial. For instance, if the probability of flipping a head on a fair coin is 0.5, that probability must remain constant for each coin flip.

# Given these conditions, the binomial distribution models the **number of successes** in \( n \) independent trials, where the probability of success on each trial is \( p \) and the probability of failure is \( 1 - p \).

### The Binomial Distribution Formula

# The probability of observing exactly \( k \) successes in \( n \) independent trials, each with a probability \( p \) of success, is given by the **binomial probability mass function**:

# P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}


# Where:
# - \( X \) is the number of successes,
# - \( n \) is the total number of trials,
# - \( k \) is the number of successes,
# - \( p \) is the probability of success on each trial,
# - \( \binom{n}{k} \) is the binomial coefficient, representing the number of ways to choose \( k \) successes from \( n \) trials.

### Relationship to the Bernoulli Distribution

# The **binomial distribution** is a **generalization** of the **Bernoulli distribution**. Here's how they relate:

# - **Bernoulli Distribution**:
#  - The Bernoulli distribution describes the probability of a **single trial** with two possible outcomes (success or failure). It models the result of one experiment (e.g., flipping a coin once).
#  - The Bernoulli distribution has **one parameter**, \( p \), the probability of success in a single trial.

# - **Binomial Distribution**:
#  - The binomial distribution models the number of successes in **multiple trials** (a fixed number of trials, \( n \)). Each trial is independent, and the probability of success is \( p \) for each trial.
#  - The binomial distribution has **two parameters**: \( n \) (the number of trials) and \( p \) (the probability of success in each trial).

### How the Binomial Distribution Relates to the Bernoulli Distribution

# The binomial distribution is essentially a series of **independent Bernoulli trials**. If you have **one trial** (i.e., \( n = 1 \)), the binomial distribution reduces to the Bernoulli distribution.

# - In other words, **if \( n = 1 \)** in the binomial distribution, the binomial probability formula simplifies to the Bernoulli probability formula:


#   P(X = 1) = p \quad \text{(success)}
#   P(X = 0) = 1 - p \quad \text{(failure)}


# Thus, a **single Bernoulli trial** is a **binomial distribution with \( n = 1 \)**.

```

```python
# Q5. What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?

### Key Properties of the Poisson Distribution

# The **Poisson distribution** is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space, given certain conditions. It is particularly useful when the events are rare or random, and they happen independently of each other.

# Here are the **key properties** of the Poisson distribution:

# 1. **Discrete Distribution**:
#   - The Poisson distribution describes the probability of **counting events** (such as the number of arrivals, occurrences, or defects) within a fixed period or region. The outcomes are non-negative integers (\( k = 0, 1, 2, \dots \)).

# 2. **Parameter**:
#   - The Poisson distribution is characterized by a single parameter \( \lambda \), which represents the **average number of events** that occur within the fixed interval. Mathematically:
#      \lambda = \text{mean number of events in the interval}

#   - The parameter \( \lambda \) must be greater than or equal to zero (\( \lambda \geq 0 \)).

# 3. **Probability Mass Function (PMF)**:
#   - The probability of observing exactly \( k \) events (where \( k \) is a non-negative integer) in the fixed interval is given by the following formula:

#     P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}

#     where:
#     - \( X \) is the random variable representing the number of events,
#     - \( \lambda \) is the average number of events in the interval,
#     - \( e \) is Euler's number (approximately 2.718),
#     - \( k! \) is the factorial of \( k \).

# 4. **Mean and Variance**:
#  - Both the **mean** and **variance** of a Poisson distribution are equal to \( \lambda \). That is:

#     E(X) = \lambda

#     \text{Var}(X) = \lambda

#   - This property implies that, as the average number of events increases, both the expected number of events and the variability around that number also increase.

# 5. **Skewness**:
#   - The Poisson distribution is **skewed** for small values of \( \lambda \) (when the average number of events is small). As \( \lambda \) increases, the distribution approaches a **normal distribution** due to the **central limit theorem**.

# 6. **No Upper Bound**:
#   - There is no upper limit to the number of events; the distribution allows for any non-negative integer value of \( k \), although probabilities decrease as \( k \) increases.

# 7. **Rare Events**:
#   - The Poisson distribution is often used to model **rare events** that occur independently in a fixed interval. The events can occur at irregular intervals, but their occurrences are statistically independent of one another.

### When to Use the Poisson Distribution

# The Poisson distribution is appropriate in the following scenarios:

# 1. **Fixed Interval of Time or Space**:
#    - The Poisson distribution is used when the events are being counted within a fixed interval, such as:
#     - A fixed time period (e.g., the number of phone calls received in an hour).
#      - A fixed area or volume (e.g., the number of cars passing through a checkpoint in a day or the number of defects in a batch of products).

# 2. **Rare or Low-Probability Events**:
#   - The Poisson distribution is often used to model rare events that occur infrequently but can happen at any point within the interval. For example:
#     - The number of accidents occurring at an intersection in a month.
#     - The number of emails arriving at a helpdesk during a specific time frame.
#     - The number of mutations in a strand of DNA during a specific observation period.

# 3. **Events are Independent**:
#    - The events must occur **independently** of each other. The occurrence of one event does not influence the likelihood of another event occurring. For instance, the arrival of customers at a store should be independent of each other.

# 4. **Constant Rate of Occurrence**:
#   - The Poisson distribution assumes that the rate of occurrence \( \lambda \) is **constant** over the interval. That is, events should occur at a steady average rate, not in bursts or at changing rates.

### Examples of When the Poisson Distribution is Used:

# - The number of customer arrivals at a service counter in an hour.
# - The number of email spam messages received in a day.
# - The number of phone calls at a call center during a specific period.
# - The number of accidents at an intersection in a year.
# - The number of occurrences of a rare disease in a population over a certain period.
```

```python
# Q6. Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?


### 1. **Probability Distribution**

# A **probability distribution** is a mathematical function or rule that describes the likelihood or probability of different outcomes in an experiment or random process. It provides a complete description of the probabilities of all possible outcomes for a given random variable.

# - **For discrete random variables**, the probability distribution specifies the probability associated with each possible value of the random variable.
# - **For continuous random variables**, the probability distribution describes how the probabilities are spread over a range of possible values.

# There are two types of probability distributions:
# - **Discrete Probability Distribution** (e.g., for discrete random variables like the number of heads in coin flips).
# - **Continuous Probability Distribution** (e.g., for continuous random variables like heights or weights).

### 2. **Probability Density Function (PDF)**

# A **probability density function** (PDF) is a function used to describe the probability distribution of a **continuous random variable**. It specifies the likelihood of the random variable taking a particular value. However, the probability of a continuous random variable taking an exact value is always zero (since there are infinitely many possible values), so the PDF instead represents the likelihood of the variable falling within a specific interval.

# - The **area under the curve** of the PDF over a given range corresponds to the probability that the random variable falls within that range.
# - The PDF is a non-negative function, and the **total area under the curve** (i.e., the integral of the PDF over its entire range) must equal 1.

# Mathematically, the PDF is often denoted by \( f(x) \), where \( f(x) \) gives the probability density at a point \( x \). To find the probability that a continuous random variable \( X \) falls within an interval \( [a, b] \), we compute the integral of the PDF over that interval:

# P(a \leq X \leq b) = \int_a^b f(x) \, dx


### 3. **Probability Mass Function (PMF)**

# A **probability mass function** (PMF) is the discrete counterpart to the PDF, and it is used to describe the probability distribution of a **discrete random variable**. The PMF gives the probability that a discrete random variable takes a particular value. Unlike the PDF, the PMF directly assigns a probability to each possible outcome.

# - The PMF is a function \( p(x) \) where \( p(x) = P(X = x) \), which gives the probability of the random variable \( X \) being equal to a specific value \( x \).
# - The sum of all probabilities in the PMF for all possible outcomes must equal 1:


 #  \sum_x p(x) = 1


# For example, in a fair six-sided die roll, the PMF would assign the probability \( \frac{1}{6} \) to each of the six possible outcomes (1, 2, 3, 4, 5, and 6).

### Differences Between PDF and PMF

# The main difference between a **probability density function** (PDF) and a **probability mass function** (PMF) lies in the type of random variable they describe and how probabilities are represented:

# | Feature                         | Probability Mass Function (PMF)                        | Probability Density Function (PDF)                        |
# |----------------------------------|--------------------------------------------------------|-----------------------------------------------------------|
# | **Type of Random Variable**     | Discrete random variables (e.g., number of heads in coin flips, number of defective items in a batch) | Continuous random variables (e.g., height, weight, temperature) |
# | **Output**                       | Assigns probabilities to specific values (e.g., \(P(X = x)\)) | Describes the probability density over an interval (e.g., \(f(x)\)) |
# | **Probability Calculation**      | \( P(X = x) = p(x) \) where \( p(x) \) is the probability mass at \( x \) | \( P(a \leq X \leq b) = \int_a^b f(x) dx \), probability is the area under the curve |
# | **Total Probability**            | Sum of all probabilities equals 1: \( \sum p(x) = 1 \) | Integral of the PDF over the entire range equals 1: \( \int_{-\infty}^{\infty} f(x) dx = 1 \) |
# | **Range of Values**              | Defined for countable outcomes (e.g., integers, distinct values) | Defined for a continuous range of values (e.g., any real number in an interval) |

```

```python
# Q7. Explain the Central Limit Theorem (CLT) with example.

### **Central Limit Theorem (CLT)**

# The **Central Limit Theorem** (CLT) is one of the most important results in statistics. It describes how the distribution of the sample mean of a large number of independent, identically distributed (i.i.d.) random variables approaches a **normal distribution** (Gaussian distribution), regardless of the original distribution of the data, as the sample size increases.

#### **Statement of the Central Limit Theorem:**

# If \( X_1, X_2, ..., X_n \) are independent random variables, each with the same distribution, mean \( \mu \), and variance \( \sigma^2 \), then the sampling distribution of the sample mean \( \overline{X} \) approaches a normal distribution as the sample size \( n \) becomes large. Specifically, the CLT states that:


# \frac{\overline{X} - \mu}{\sigma/\sqrt{n}} \xrightarrow{n \to \infty} N(0,1)


# Where:
# - \( \overline{X} \) is the sample mean.
# - \( \mu \) is the population mean.
# - \( \sigma \) is the population standard deviation.
# - \( n \) is the sample size.
# - \( N(0,1) \) denotes the standard normal distribution (mean = 0, standard deviation = 1).

# This means that, as \( n \) grows, the distribution of the sample mean \( \overline{X} \) becomes approximately normal with a mean \( \mu \) and a standard deviation \( \frac{\sigma}{\sqrt{n}} \), even if the original data are not normally distributed.

#### **Key Concepts in the CLT:**

# 1. **Sample Mean Distribution**:
#   - As \( n \) increases, the **distribution of the sample mean** becomes more and more like a normal distribution, even if the original population distribution is not normal.
#   - The **mean** of the sample mean is \( \mu \) (the population mean).
#   - The **standard deviation** of the sample mean (also called the standard error) is \( \frac{\sigma}{\sqrt{n}} \), which decreases as \( n \) increases.

# 2. **Size of the Sample**:
#   - The larger the sample size \( n \), the more closely the sampling distribution of the sample mean will approximate a normal distribution. Typically, a sample size of \( n \geq 30 \) is considered large enough for the CLT to apply, but in some cases, smaller sample sizes might suffice, depending on the shape of the original distribution.

# 3. **Independence**:
#   - The random variables (samples) must be independent of each other. This means that the outcome of one trial should not affect the outcome of another.

#### **Example of the Central Limit Theorem:**

# Let's consider an example to better understand the CLT:

# **Scenario**: Suppose we are interested in the average height of adult women in a country. The distribution of heights in this country is not perfectly normal (it could be skewed or follow some other shape), but we still want to estimate the average height from a sample of women.

# - **Step 1**: We randomly select a sample of **30 women** (which is typically considered large enough for the CLT to apply).
# - **Step 2**: We calculate the **sample mean height** and the **sample standard deviation** for that sample.
# - **Step 3**: Now, imagine we repeat this process â€” selecting a random sample of 30 women, calculating the mean height, and repeating this process many times (say, 10,000 times).
# - **Step 4**: According to the Central Limit Theorem, if we plot all these sample means (i.e., the distribution of sample means), it will **approach a normal distribution**. This happens even if the original distribution of heights is not normal (e.g., if the original distribution is skewed).

# **Why this is useful**: The CLT allows us to make inferences about the population mean (in this case, the average height of all adult women in the country) based on the sample mean, even when the original data is not normally distributed. Since the distribution of sample means will be approximately normal, we can use standard techniques from normal distribution (like confidence intervals or hypothesis tests) to estimate population parameters.

#### **Illustration**:
# - Suppose the original distribution of heights has a **mean** of 160 cm and a **standard deviation** of 10 cm.
# - If we repeatedly take samples of size 30, the distribution of the sample means will have:
#   - Mean: 160 cm (same as the population mean).
#   - Standard deviation (standard error): \( \frac{10}{\sqrt{30}} \approx 1.83 \) cm.

# Even if the original distribution of heights is not normal, the sample means will approach a normal distribution as we increase the number of samples.

#### **Why the CLT is Important:**

# 1. **Making Inferences**: It allows us to use the normal distribution to make statistical inferences (like confidence intervals and hypothesis tests) even when the data are not normally distributed.
# 2. **Simplifies Calculations**: The normal distribution is well-understood and has many properties that make statistical analysis easier.
# 3. **Wide Applicability**: The CLT applies to a wide variety of problems in different fields like economics, biology, engineering, and social sciences.

```

```python
# Q8.  Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?


### **Comparison of Z-scores and T-scores**

# Both **z-scores** and **t-scores** are statistical measures used to describe the position of a data point or sample mean relative to a population, but they are used in different contexts and have some key differences. Here's a comparison:

# | **Aspect**                   | **Z-score**                             | **T-score**                             |
# |------------------------------|-----------------------------------------|-----------------------------------------|
# | **Definition**                | The **z-score** measures how many **standard deviations** a data point or sample mean is from the population mean. It is used when the population standard deviation is known. | The **t-score** is similar to a z-score but is used when the population standard deviation is **unknown** and the sample size is small. It accounts for extra uncertainty introduced by estimating the population standard deviation. |
# | **Formula**                   | \( z = \frac{X - \mu}{\sigma} \)          | \( t = \frac{\overline{X} - \mu}{s / \sqrt{n}} \) |
# | **Variables**                 | \( X \) = individual data point, \( \mu \) = population mean, \( \sigma \) = population standard deviation | \( \overline{X} \) = sample mean, \( \mu \) = population mean, \( s \) = sample standard deviation, \( n \) = sample size |
# | **Population Standard Deviation** | Used when the population standard deviation (\( \sigma \)) is **known** | Used when the population standard deviation is **unknown**, and sample standard deviation \( s \) is used instead |
# | **Sample Size**               | Typically used for **large samples** (typically \( n > 30 \)) or when the population is large enough to make approximations | Used for **small samples** (typically \( n \leq 30 \)) or when the sample size is small enough to require an estimate of the population standard deviation |
# | **Distribution**               | The z-score follows the **standard normal distribution**, which is symmetrical with a mean of 0 and a standard deviation of 1. | The t-score follows the **t-distribution**, which is similar to the normal distribution but has thicker tails, especially for small sample sizes. As the sample size increases, the t-distribution approaches the normal distribution. |
# | **Application**                | Used for hypothesis testing, confidence intervals, and comparing a single data point or sample mean to a population when the population parameters are known. | Used for hypothesis testing, confidence intervals, and comparing sample means to a population mean when the population standard deviation is unknown and/or the sample size is small. |

### **When to Use a Z-score**

# You should use a **z-score** when:
# 1. **The population standard deviation is known**: This is crucial because the z-score formula requires the population standard deviation (\( \sigma \)).
# 2. **You have a large sample size**: Generally, when \( n > 30 \), the **Central Limit Theorem (CLT)** ensures that the sampling distribution of the sample mean will be approximately normal, even if the underlying data is not normally distributed. In such cases, the z-score is appropriate.
# 3. **For comparing individual data points to a population mean**: You can use z-scores to see how far a specific value is from the mean in terms of standard deviations.

# **Example**:
# - You are testing whether a studentâ€™s score of 85 on a test is significantly different from the population mean score of 75, and you know the population standard deviation is 10. You can calculate the z-score to assess how far the studentâ€™s score is from the population mean.

### **When to Use a T-score**

# You should use a **t-score** when:
# 1. **The population standard deviation is unknown**: When you don't know the population standard deviation and must estimate it using the sample standard deviation (\( s \)), the t-score should be used.
# 2. **The sample size is small**: For small sample sizes (\( n \leq 30 \)), the t-distribution is used because it accounts for more variability when estimating population parameters from smaller samples.
# 3. **For small samples or confidence intervals for sample means**: When the sample size is small and you are estimating the mean of a population, the t-distribution is more appropriate than the normal distribution, because it accounts for the added uncertainty in the estimate of the population standard deviation.

# **Example**:
# - You are conducting a survey of 20 people to estimate the average number of hours they spend on social media each day. Since you donâ€™t know the population standard deviation, you would use the t-distribution to calculate a confidence interval for the population mean.

### **Key Differences Between Z-scores and T-scores**:

# 1. **Population Standard Deviation**: Z-scores are used when the population standard deviation is known, while t-scores are used when it is unknown.

# 2. **Sample Size**: Z-scores are typically used for larger sample sizes (\( n > 30 \)), while t-scores are used for smaller sample sizes (\( n \leq 30 \)).

# 3. **Distribution Shape**: Z-scores follow the **standard normal distribution**, which is symmetrical and has a mean of 0 and standard deviation of 1. T-scores follow the **t-distribution**, which is also symmetrical but has thicker tails, especially for smaller sample sizes. As the sample size increases, the t-distribution approaches the normal distribution.

# 4. **Tail Thickness**: The t-distribution has thicker tails, which reflect greater uncertainty when estimating the population parameters from small samples. This means that t-scores are generally more conservative, making it harder to reject the null hypothesis when the sample size is small.
```

```python
# Q9.  Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?
# Task: Write Python code to calculate the z-score and p-value for the given data.
# Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing


# To solve this problem, we need to calculate the z-score and p-value, and then make a decision about whether to reject or fail to reject the null hypothesis based on a significance level of 0.05.

### Step-by-step Solution:

#### 1. **Calculate the z-score:**

# The formula for the z-score is:

# z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}

# Where:
# - \(\bar{x}\) = Sample mean = 105
# - \(\mu\) = Population mean = 100
# - \(\sigma\) = Standard deviation = 15
# - \(n\) = Sample size = 25

#### 2. **Calculate the p-value:**

# Once we calculate the z-score, the p-value is found by using the cumulative distribution function (CDF) for the z-score. This gives the probability of observing a value less than or equal to the given z-score in a standard normal distribution.

#### 3. **Decision Rule:**

# - If the p-value is less than the significance level (0.05), reject the null hypothesis.
# - If the p-value is greater than or equal to the significance level, fail to reject the null hypothesis.

### Python Code to Calculate Z-Score and P-Value:

import scipy.stats as stats
import math

sample_mean = 105
population_mean = 100
std_deviation = 15
sample_size = 25
alpha = 0.05

z_score = (sample_mean - population_mean) / (std_deviation / math.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

z_score, p_value, result
```
